```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data_set_url = 'training_set/output_final.csv'
training_set = pd.read_csv(data_set_url, header=0, sep=',') 
training_set.head()
```

```python
header = ['red_pixel','green_pixel','blue_pixel','mean_red','mean_green','mean_blue','std_red','std_green','std_blue','corr_rg','corr_rb','corr_gb','inside']
```

```python
test_set_url = 'test_set/output_test_set.csv'
test_set = pd.read_csv(test_set_url, header=0, sep=',') 
test_set.head()
```

```python
test_set.describe()
```

```python
training_set[header] = training_set[header].astype(float)
training_set.dtypes
```

```python
training_set.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(training_set)
```

```python
training_set[header] = scaler.fit_transform(training_set[header])
```

```python
training_set.describe()
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
y_train = targets = training_set['inside'] # dependent variable
x_train = training_set.drop(['inside'],axis=1)
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
test_set[header] = test_set[header].astype(float)
test_set.dtypes
```

```python
scaler = MinMaxScaler()
scaler.fit(test_set)
test_set[header] = scaler.fit_transform(test_set[header])
```

```python
y_test = test_set['inside'] # dependent variable
x_test = test_set.drop(['inside'],axis=1)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm)
cm_df
cm_df.columns = ['1','0']
cm_df = cm_df.rename(index={0: '1',1:'0'})
cm_df
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
import pickle
model_file_name = 'naive_bayes_classifier_model.pkl'
file = open(model_file_name, 'wb')
pickle.dump(classifier, file)
file.close()
```

```python
loaded_model = pickle.load(open('naive_bayes_classifier_model.pkl', 'rb'))
```

```python
from sklearn.metrics import confusion_matrix
y_pred = loaded_model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm)
cm_df
cm_df.columns = ['1','0']
cm_df = cm_df.rename(index={0: '1',1:'0'})
cm_df
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
y_train = targets = training_set['inside'] # dependent variable
x_train = training_set.drop(['inside'],axis=1)
```

```python
classifier = RandomForestClassifier(verbose=3)
```

```python
classifier.fit(x_train,y_train)
```

```python
test_set[header] = test_set[header].astype(float)
test_set.dtypes
```

```python
scaler = MinMaxScaler()
scaler.fit(test_set)
test_set[header] = scaler.fit_transform(test_set[header])
```

```python
y_test = test_set['inside'] # dependent variable
x_test = test_set.drop(['inside'],axis=1)
```

```python
from sklearn.metrics import confusion_matrix
y_pred = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm)
cm_df
cm_df.columns = ['1','0']
cm_df = cm_df.rename(index={0: '1',1:'0'})
cm_df
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
